```python
#Sistema de dialogo dirigido a tareas para tomar una orden de café

#Estado inicial del dialogo

orden = {
    'tipo' : None,
    'tamaño': None,
    'temperatura': None,
    'endulzantes': None,
    'añadidos': None,
    'alimentos': None,
    'propina': None,
    'leche': None
}

def pedir_tamaño():
  tamaño = input('¿Que tamaño deseas? (Pequeño, Mediano, Grande):').lower()
  if tamaño in ['pequeño', 'mediano', 'grande']:
    orden['tamaño'] = tamaño
    return True
  else:
    print("Lo siento, no entendi eso. Puedes elegir Pequeño, Mediano o Grande")


def pedir_tipo():
    tipo = input('¿Qué tipo de café deseas? (Espresso, Americano, Capuchino, latte , mocha, frappe): ').lower()
    if tipo in ['espresso', 'americano', 'capuchino', 'latte', 'mocha', 'frappe']:
        orden['tipo'] = tipo
        return True

def tipo_de_leche():
    leche = input('¿Qué tipo de leche deseas? (Entera, Deslactosada, Light, Almendras, Soya,ninguna): ').lower()
    if leche in ['entera', 'deslactosada', 'light', 'almendras', 'soya', 'ninguna']:
        orden['leche'] = leche
        return True
    

def pedir_temperatura():
    temperatura = input('¿Cómo deseas la temperatura de tu café? (Caliente, Templado, Frío): ').lower()
    if temperatura in ['caliente', 'templado', 'frío']:
        orden['temperatura'] = temperatura
        return True

def pedir_endulzantes():
    endulzante = input('¿Quieres añadir endulzantes? (Ninguno, Azúcar, Stevia, miel ,splenda): ').lower()
    if endulzante in ['ninguno', 'azúcar', 'stevia', 'miel', 'splenda']: 
        orden['endulzantes'] = endulzante
        return True

def pedir_añadidos():
    añadido = input('¿Deseas añadir algo más a tu café? (Crema, Canela, chocolate, vainilla): ').lower()
    if añadido in ['crema', 'canela', 'chocolate', 'vainilla']:
        orden['añadidos'] = añadido
        return True

def pedir_alimentos():
    alimento = input('¿Quieres algo de comer? (Galletas, Pastel, Ninguno): ').lower()
    if alimento in ['galletas', 'pastel', 'ninguno']:
        orden['alimentos'] = alimento
        return True

def pedir_propina():
    propina = input('¿Cuánto te gustaría dejar de propina? (Ingresa un número): ')
    try:
        propina = float(propina)
        orden['propina'] = propina
    except ValueError:
        print("Por favor, ingresa un número válido para la propina.")




 
def realizar_pedido():
    print("☕️----BIENVENIDO A COFFE STATION ----☕️")
    pedir_tamaño()
    pedir_tipo()
    pedir_temperatura()
    pedir_endulzantes()
    pedir_añadidos()
    pedir_alimentos()
    pedir_propina()
    print("Tu pedido es:", orden)
    print("☕️----GRACIAS X SU PREFERENCIA ----☕️")


realizar_pedido()

  
```

```python
import random
import datetime

# Simulación de la base de datos con información de diferentes temas
info_db = {
    "clima": [
        "Hoy el clima está soleado con una ligera brisa",
        "Se espera que el clima esté nublado con una temperatura de 20 grados",
        "Está bastante frío, se espera que la temperatura baje a 10 grados",
    ],
    "definiciones": {
        "python": "Python es un lenguaje de programación interpretado cuya filosofía hace hincapié en la legibilidad de su código",
        "inteligencia artificial": "La inteligencia artificial es la inteligencia llevada a cabo por máquinas",
        "chatbot": "Un chatbot es un programa que intenta simular un ser humano en una conversación",
        "base de datos": "Una base de datos es un conjunto de datos pertenecientes a un mismo contexto y almacenados sistemáticamente para su posterior uso",
        "algoritmo": "Un algoritmo es un conjunto finito de instrucciones que, siguiendo un orden determinado, realiza una tarea específica",
    }
}

# Función que maneja el diálogo y busca la respuesta en la base de datos simulada
def responder_pregunta(tema, detalle=None):
    if tema == 'clima':
        return random.choice(info_db['clima'])
    elif tema == 'hora':
        return "Son las {} con {} minutos".format(datetime.datetime.now().hour, datetime.datetime.now().minute)
    elif tema == 'definiciones' and detalle:
        return info_db['definiciones'].get(detalle, "Lo siento, no tengo esa definición")
    else:
        return "Lo siento, no entendí eso. Puedes preguntar sobre el clima, la hora o definiciones."

# Función principal del sistema de diálogo
def sistema_dialogo_informativo():
    print("Bienvenido al sistema de diálogo informativo, pregunta sobre el clima, la hora o definiciones de conceptos específicos.")

    activo = True
    while activo:
        pregunta = input("¿Qué deseas saber? ").lower()
        if 'clima' in pregunta:
            print(responder_pregunta('clima'))
        elif 'hora' in pregunta:
            print(responder_pregunta('hora'))
        elif 'que es' in pregunta:
            palabras = pregunta.split()
            indice = palabras.index('es') + 1
            detalle = ' '.join(palabras[indice:]) if indice < len(palabras) else None
            print(responder_pregunta('definiciones', detalle))
        elif 'adios' in pregunta:
            print("Hasta luego.")
            activo = False
        else:
            print("No entendí eso, puedes preguntar sobre el clima, la hora o definiciones.")

sistema_dialogo_informativo()
```
