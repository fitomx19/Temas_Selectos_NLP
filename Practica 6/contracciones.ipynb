```python
diccionario = {
    "don't": "do not",
    "doesn't": "does not",
    "can't": "cannot",
    "I'm": "I am",
    "you're": "you are",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "we're": "we are",
    "they're": "they are",
    "I'll": "I will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "it'll": "it will",
    "we'll": "we will",
    "they'll": "they will",
    "I've": "I have",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "I'd": "I would",
    "you'd": "you would",
    "he'd": "he would",
    "she'd": "she would",
    "it'd": "it would",
    "we'd": "we would",
    "they'd": "they would",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "hasn't": "has not",
    "haven't": "have not",
    "hadn't": "had not",
    "won't": "will not",
    "wouldn't": "would not",
    "couldn't": "could not",
    "shouldn't": "should not",
    "mightn't": "might not",
    "mustn't": "must not",
    "shan't": "shall not",   
}
```

```python

def expandir(texto):
    for contraction, expansion in diccionario.items():
        texto = texto.replace(contraction, expansion)
    return texto

def contraer(texto):
    for contraction, expansion in diccionario.items():
        texto = texto.replace(expansion, contraction)
    return texto
```

```python
opcion = input("¿Quieres contraer o expandir las palabras? (c/e): ")

if opcion.lower() == "c":
    resultado = contraer(parrafo)
elif opcion.lower() == "e":
    resultado = expandir(parrafo)
else:
    resultado = "Opción no válida. Por favor, elige 'contraer' o 'expandir'."

print("With or Without You (U2)")
print("----------------------------------------")
print("Resultado:")
print(resultado)
```
