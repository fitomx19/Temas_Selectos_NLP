Importar ambos archivos

```python
human_text = open("human_text.txt", "r")
robot_text = open("robot_text.txt", "r")
```

```python
human_text = human_text.read()
robot_text = robot_text.read()
```

Visualizamos el contenido

```python
print("Human text: \n", human_text)
```

Primero realizamos una correccion de ortografia simple

```python
def correct_text(text):
    spell = SpellChecker()
    text = text.lower()
    tokens = word_tokenize(text)
    corrected_text = []
    for word in tokens:
        corrected_text.append(spell.correction(word))
    return corrected_text
```

Funcion para normalizar el texto

```python
def normalize_text(text):
    #Pasamos a minusculas las palabras
    text = text.lower()
    #Eliminamos los signos de puntuacion
    text = text.translate(str.maketrans('', '', string.punctuation))
    #Eliminamos los espacios en blanco
    text = text.strip()
    #eliminamos los emojis
    text = text.encode('ascii', 'ignore').decode('ascii')
    return text
```

Eliminar __stopwords__

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    
    return filtered_text
```

__Lemmatizar__ el texto

```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    lemmatized_sentence = [lemmatizer.lemmatize(w) for w in text]
    return lemmatized_sentence
```

Creamos __"bag of words"__

```python
def bolsa_de_palabras(texto):
    texto = normalize_text(texto)
    texto = remove_stopwords(texto)
    texto = lemmatize_text(texto)
     
    return texto
```

Obtenemos la bolsa de palabras de cada texto 

```python
human_text_bolsa = bolsa_de_palabras(human_text)
human_text_str = ' '.join(human_text_bolsa)
print("Human text bolsa de palabras: \n", human_text_bolsa)

robot_text_bolsa = bolsa_de_palabras(robot_text)
robot_text_str = ' '.join(robot_text_bolsa)
print("Robot text bolsa de palabras: \n", robot_text_bolsa)
```

```python
def obtener_frecuencia_palabras(texto):
    obtener_frecuencia = {}
    for palabra in texto:
        if palabra not in obtener_frecuencia.keys():
            obtener_frecuencia[palabra] = 1
        else:
            obtener_frecuencia[palabra] += 1
    return obtener_frecuencia
```

```python
frecuencia_palabras_human_text = obtener_frecuencia_palabras(human_text_bolsa)
print("Frecuencia de palabras human text: \n", frecuencia_palabras_human_text)
```

```python
frecuencia_palabras_robot_text = obtener_frecuencia_palabras(robot_text_bolsa)
print("Frecuencia de palabras robot text: \n", frecuencia_palabras_robot_text)
```

------------

"__Bag of words__ de ambos textos"

```python
bolsa_de_palabras_grandotota = human_text_bolsa + robot_text_bolsa

frecuencia_palabras_bolsota = obtener_frecuencia_palabras(bolsa_de_palabras_grandotota)
```
